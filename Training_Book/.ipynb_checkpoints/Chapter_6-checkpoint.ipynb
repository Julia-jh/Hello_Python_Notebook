## 6. 2. 2. 데이터의 결합

```python
# 데이터 1 생성
data1 = {
    'id': ['100', '101', '102', '103', '104', '106', '108', '110', '111', '113'],
    'city': ['Seoul', 'Pusan', 'Daegu', 'Gangneung', 'Seoul', 'Seoul', 'Pusan', 'Daegu', 'Gangneung', 'Seoul'],
    'birth_year': [1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
    'name': ['Junho', 'Heejin', 'Mijung', 'Minho', 'Steeve', 'Mina', 'Sumi', 'Minsu', 'Jinhee', 'Daeho']
}

df1 = DataFrame(data1)
df1
```

```python
# 데이터 2 생성
data2 = {
    'id': ['100', '101', '102', '105', '107'],
    'math': [50, 43, 33, 76, 98],
    'English': [90, 30, 20, 50, 30],
    'sex': ['M', 'F', 'F', 'M', 'M'],
    'index_num': [0, 1, 2, 3, 4]
}

df2 = DataFrame(data2)
df2
```

```python
# 내부 결합

# 데이터 결합(내부결합. 키는 자동으로 인식하지만 on으로 명시적으로 지정할 수 있음)
print('.결합 테이블')
pd.merge(df1, df2, on = 'id')
```

```python
# 외부 결합

# 데이터 결합(전결합)
pd.merge(df1, df2, how = 'outer')
```

```python
# index를 이용한 결합-num
pd.merge(df1, df2, left_index = True, right_on = 'index_num')
```

### 1. left_index와 right_on 값을 이것저것 바꿔보기

### 1의 해설

- how : join 방법 (default는 inner join), 

         - Outer Join은 'outer' / Inner Join은 'inner' / Left Join은  'left' / Right Join은 'right'

- on : join할 key (같은 key 값을 갖을 경우)

         - join할 키가 다수인 경우 리스트 형식으로 전달

- left_on, right_on : 각 join할 key (서로 다른 key 값을 갖을 경우)
- left_index, right_index : join할 key가 인덱스일 경우
- sort : 조인 키 정렬



출처: https://data-make.tistory.com/129 [Data Makes Our Future]

```python
pd.merge(df1, df2, left_index = True, right_index = True)
```

```python
# 왼쪽 결합

# 데이터 결합(left)
pd.merge(df1, df2, how = 'left')
```

- 1달 단위 구간    : resample('1M')

- 1년 단위 구간    : resample('1Y')
